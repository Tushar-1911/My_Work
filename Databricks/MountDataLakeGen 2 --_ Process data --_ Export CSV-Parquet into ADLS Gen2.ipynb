###Data Cleaning

######Fill null values as 'NA'

```
df =df.fillna('NA')
display(df)
```

######Filter out zero values from Passenger_count and trip_distance columns

```
filtered_df = df.filter("passenger_count > 0")

filtered_df = df.where("trip_distance > 0")
```

```
display(filtered_df)
```

```
display(
    filtered_df.describe(
                                     "passenger_count",                                     
                                     "trip_distance"                                     
                                 )
)
```

###### Drop NA values from PULocationID and DOLocationID columns

```
filtered_df = filtered_df.dropna(thresh=2,subset=("PULocationID", "DOLocationID"))
```

### Transformation

###### Rename the columns

```
filtered_df = filtered_df.withColumnRenamed("PULocationID","PickupLocationId").withColumnRenamed("DOLocationID","DropLocationId")
```

```
display(filtered_df)
```

###Loading data

###### Load CSV in ADLS Gen2

```
filtered_df.write.mode("overwrite").format("com.databricks.spark.csv").option("header","true").csv("abfs://taxi@adlastg.dfs.core.windows.net/imports/taxi_data.csv")
```

###### Load Parquet in ADLS Gen2

```
filtered_df.write.mode("overwrite").format("com.databricks.spark.csv").option("header","true").parquet("abfs://taxi@adlastg.dfs.core.windows.net/imports/taxi_data.parquet")
```
